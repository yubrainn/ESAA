## for문

```python
# 구구단(2739)

N = int(input())
for i in range(1,10):
    print(N,'*',i,'=',N*i)
    
```

```python
# A+B-3
T = int(input())

for i in range(T):
    a, b = map(int, input().split())
    print(a+b)
```

```python
# 합
n = int(input())
sum = 0
for i in range(1, n+1):
    sum += i
print(sum)
```

```python
# N 찍기
N = int(input())
for i in range(1,N+1):
    print(i)
```

```python
# 기찍 N
N = int(input())
for i in range(N):
    print(N-i)
```

```python
# A+B-7
T = int(input())

for c in range(1,T+1):
    A, B = map(int, input().split())
    print("Case #{0}:".format(c), A+B)
```

```python
# A+B-8
T = int(input())

for c in range(1,T+1):
    A, B = map(int, input().split())
    print("Case #{0}: {1} + {2} =".format(c,A,B), A+B)
```

```python
# 별 찍기 - 1
N = int(input())

for i in range(1, N+1):
    print("*"*i)
```

```python
# 별 찍기 - 2
N = int(input())
for i in range(1, N+1):
    print(' '*(N-i)+"*"*i)
```

## while문

```python
# A+B-5
check = [0, 0]
tc = []
while True:
    tc = list(map(int, input().split()))
    if tc == check: break
    print(tc[0]+tc[1])
```

```python
# 더하기 사이클
N = int(input())
cy = 0
new = N
while True:
    ten = new //10
    one = new % 10
    to = ten + one
    new = one * 10 + to % 10
    cy +=1
    if N == new : break
    
    
print(cy)
```
