```python
import warnings
warnings.filterwarnings("ignore")

%matplotlib notebook
import matplotlib.pyplot as plt

import seaborn as sns
```

### 2.1 카테고리 - 투표 수 - 청원 시기
어떤 시기, 어떤 카테고리의 글이 투표를 많이 받았는지에 대해 분석

**카테고리별 투표 수 시각화**

```python
#카테고리 별 투표 수
cat_votes = df.groupby('category',as_index=False).votes.sum()
cat_votes
```

```python
plt.figure(figsize=(20,7))
plt.rcParams['font.family'] = 'NanumGothic'
ax = sns.barplot(x = 'category', y = 'votes', data = cat_votes)
```

인권/성평등 항목의 투표 수가 가장 높은 것을 알 수 있다.  
이는 총 투표수이므로 카테고리에서 한 청원에 대한 평균 투표 수를 알아보기 위해 새로운 그래프를 그려보자

**카테고리 별 평균 투표 수**

```python
#카테고리 별 평균투표 수
cat_votes_m = df.groupby('category',as_index=False).votes.mean()
cat_votes_m
```

```python
plt.figure(figsize=(20,7))
plt.rcParams['font.family'] = 'NanumGothic'
ax = sns.barplot(x = 'category', y = 'votes', data = cat_votes_m)
```

각 청원에 대한 평균 투표 수는 반려동물이 가장 많았고, 그 다음이 인권/성평등임을 볼 수 있다.  
이제 청원 시기와 카테고리별 투표 수를 합쳐, 어느 시기에 어떤 카테고리의 청원이 많은 투표를 받았는지 알아보자

```python
# 청원 시작 날짜를 기준으로 년,월만 뽑아 새로운 열 생성
import datetime as dt

df['year'] = df['start'].dt.strftime('%Y')
df['month'] = df['start'].dt.strftime('%m')
df['yyyy-mm'] = df['start'].dt.strftime('%Y-%m')
```

```python
#카테고리 별 투표 수 + 시기
ym_cat_votes = df.groupby(['category', 'month',  'year'],as_index=False).votes.sum()
ym_cat_votes
```

```python
facet = sns.FacetGrid(ym_cat_votes, col = 'category', hue = 'year', height = 8)
facet.map(sns.lineplot, 'month', 'votes')
facet = facet.add_legend()
```

```python
#카테고리 별 투표 수 + 연-월 시기
y_m_cat_votes = df.groupby(['category','yyyy-mm'], as_index=False).votes.sum()
y_m_cat_votes
```

```python
facet = sns.FacetGrid(y_m_cat_votes, col = 'category', height = 8)
facet.map(sns.lineplot, 'yyyy-mm', 'votes')
facet.map(lambda y, **kw: plt.axhline(y.mean(), linewidth = 1, color = 'r' ), 'votes')
facet = facet.add_legend()
```

시기와 카테고리별 투표 수를 보았을 때 농산어촌, 일자리, 저출산/고령화대책은 일관적으로 투표율이 낮다.  
대부분은 평균을 중심으로 등락이 있는 편이고, 미래, 성장동력은 한 시점에서만 높은 투표 수를 보인다.

### 2.2 카테고리 별 답변대상 청원 수
어떤 카테고리에서 답변 대상 청원이 많이 나왔는지 알아보자

```python
# 답변 대상 청원만 뽑아서 보기
answer_obj = df[df['answer'] == 1]
answer_obj
```

```python
# 히스토그램
plt.figure(figsize=(15, 5))
ax = sns.histplot(data = answer_obj, x = 'category')
```

```python
# 산점도
ax = sns.relplot(x = 'start', y = 'votes', data = df, hue = 'category')
ax.map(plt.axhline, y = 200000, color = '.7', dashes=(2, 1))
```

답변 대상 청원의 개수는 인권/성평등이 월등하게 많다.  
답변 대상 청원 중 투표 수가 월등하게 많은 것은 안전/환경에 속한 청원이었다.

```python
# 투표 수 높은 것 내용 확인
answer_obj.sort_values(by=['votes'], axis = 0, ascending = False)
```
