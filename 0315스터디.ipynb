reshape 함수를 (-1, 1)의 형태로 사용하게 되면 원 배열이 열의 형태로 나타나게 되고, (1, -1)의 형태로 사용하게 되면 원 배열이 행의 형태로 나타남. 또, 행은 axis = 0, 열은 axis = 1로 표기함. 헷갈리면 열은 1모양이라고 생각하기,,

```python
import numpy as np
```

```python
array1 = np.zeros((4,2))

array2 = array1.reshape(-1,1)

print(array2.shape)

print(array1)
print(array2)
```

```python
array1d = np.arange(start = 1, stop = 10)
array2d = array1d.reshape(3,3)

print(array1d)
print(array2d)
```

```python
array2d[:2, 0]
```

```python
array2d[0, :2]
```
