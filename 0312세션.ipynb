## if(2884)

```python
[h,m]=list(map(int,input().split()))
if m>=45:
    print(h,m-45)
else:
    if h>0: print(h-1,m+15)
    else: print(23, m+15)
```

## for(15552)

```python
import sys

T = int(input()) #최대 1,000,000


for i in range(T):
    A,B=map(int,sys.stdin.readline().split()) # 1이상 1,000 이하
    print(A+B)
```

## for(10871)

```python
N, X = map(int, input().split())
A = list(map(int, input().split()))

for n in A:
    if n < X:
        print(n, end= ' ')
```

## while(10951)
